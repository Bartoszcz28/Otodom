```python
import numpy as np
import pandas as pd
import psycopg2
import pandas.io.sql as sqlio
import matplotlib.pyplot as plt
import pylab as pl
%matplotlib inline
from IPython.display import clear_output
```

```python
cur = psql.cursor()
sql = "select * from Warsaw_sell_full;"
dat = sqlio.read_sql_query(sql, psql)
conn = None
dat.head(50)
dat.shape
```

```python
dat = dat.drop_duplicates(subset=['link_2']) # usuwanie duplikatów 
dat.shape
```

```python
dat.head(50)
```

```python
cdf = dat[["id","price","m2","rent","number_of_floors","room_number","floor","year_of_construction"]]
cdf.head(113)
```

```python
cdf = cdf.reset_index()
```

```python
cdf.shape
```

```python
avg = cdf
avg = avg[avg.year_of_construction.between(1750, 2022, inclusive=False)]
avg = avg["year_of_construction"].astype("float").mean(axis=0)
avg 
```

```python
for year in range(len(cdf.index)):
    if 1750 < cdf.at[year,"year_of_construction"] < 2022:
        continue
    else:
        cdf.at[year,"year_of_construction"] = avg
cdf.reset_index()
    
```

```python
for p in range(len(cdf.index)):
    if 200000 > cdf.at[p,"price"] > 5000000:
        cdf = cdf.drop([p])
```

```python
cdf.head()
```

```python
for m in range(len(cdf.index)):
    if 10 > cdf.at[m,"m2"] > 600:
        cdf = cdf.drop([m])
```

```python
column = cdf["m2"]
max_value = column.max()
print(max_value)
```

```python
column = cdf["year_of_construction"]
max_value = column.min()
print(max_value)
```

```python
avg_price = cdf["price"].astype("float").mean(axis=0)
print("Average price: ", avg_price)

avg_m2 = cdf["m2"].astype("float").mean(axis=0)
print("Average m2: ", avg_m2)

avg_rent = cdf["rent"].astype("float").mean(axis=0)
print("Average rent: ", avg_rent)

avg_number_of_floors = cdf["number_of_floors"].astype("float").mean(axis=0)
print("Average number_of_floors: ", avg_number_of_floors)

avg_room_number = cdf["room_number"].astype("float").mean(axis=0)
print("Average room_number: ", avg_room_number)

avg_floor = cdf["floor"].astype("float").mean(axis=0)
print("Average floor: ", avg_floor)

avg_year_of_construction = cdf["year_of_construction"].astype("float").mean(axis=0)
print("Average year_of_construction: ", avg_year_of_construction)

cdf["price"].replace(np.nan, avg_price, inplace=True)
cdf["m2"].replace(np.nan, avg_m2, inplace=True)
cdf["rent"].replace(np.nan, avg_rent, inplace=True)
cdf["number_of_floors"].replace(np.nan, avg_number_of_floors, inplace=True)
cdf["room_number"].replace(np.nan, avg_room_number, inplace=True)
cdf["floor"].replace(np.nan, avg_floor, inplace=True)
# cdf["year_of_construction"].replace(np.nan, avg_year_of_construction, inplace=True) # do dorobienia
```

```python
cdf.head(50)
```

```python
cdf.shape
```

```python
msk = np.random.rand(len(cdf)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[["m2","rent","number_of_floors","room_number","floor","year_of_construction"]])
y = np.asanyarray(train[["price"]])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[["m2","rent","number_of_floors","room_number","floor","year_of_construction"]])
x = np.asanyarray(test[["m2","rent","number_of_floors","room_number","floor","year_of_construction"]])
y = np.asanyarray(test[["price"]])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))
# ,"year_of_construction"
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
