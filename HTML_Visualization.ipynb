```python

# Read the airline data into pandas dataframe
airline_data =  pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/airline_data.csv', 
                            encoding = "ISO-8859-1",
                            dtype={'Div1Airport': str, 'Div1TailNum': str, 
                                   'Div2Airport': str, 'Div2TailNum': str})

# Randomly sample 500 data points. Setting the random state to be 42 so that we get same result.
data = airline_data.sample(n=500, random_state=42)

# Pie Chart Creation
fig = px.pie(data, values='Flights',names='DistanceGroup')

# Create a dash application
app = dash.Dash(__name__)


app.layout = html.Div(children=[html.H1('Airline Dashboard',
                                        style={'textAlign': 'center', 'color': '#503D36', 'font-size': 40}),
                                html.P(html.P('Proportion of distance group (250 mile distance interval group) by flights.', style={'textAlign':'center', 'color': '#F57241'}),),
                                dcc.Graph(figure=fig),

                    ])

# Run the application                   
if __name__ == '__main__':
    app.run_server(host='0.0.0.0')
```

```python
# Add computation to callback function and return graph
def get_graph():

    fig = go.Figure(data=go.Scatter(x=sm_table['rent_price'], y=sm_table['m2'], mode='markers', marker=dict(color='green')))
    fig.update_layout(title='Rent price & Flat size', xaxis_title='Rent price', yaxis_title='Flat size')
    return fig
def get_graph_1():

    fig = go.Figure(data=go.Scatter(x=dep_table['rent_price'], y=dep_table['deposit'], mode='markers', marker=dict(color='red')))
    fig.update_layout(title='Rent price & Deposit', xaxis_title='Rent price', yaxis_title='Deposit')
    return fig
def get_graph_2():

    fig = go.Figure(data=go.Scatter(x=room_table['rent_price'], y=room_table['room_number'], mode='markers', marker=dict(color='blue')))
    fig.update_layout(title='Rent price & Room Number', xaxis_title='Rent price', yaxis_title='Room Number')
    return fig

def get_graph_3():
    fig = px.pie(df, values='Number', names='Building_type', title='Distance group proportion by flights')
```

```python

# Create a dash application
app = dash.Dash(__name__)

app.layout = html.Div(children=[html.H1("Correlation between rent price and flat size",style={'textAlign': 'center', 'color': '#503D36', 'font-size': 40}),
                                html.Br(),
                                html.Br(),
                                html.Div(dcc.Graph(figure=get_graph())),
                                html.Br(),
                                html.Div(dcc.Graph(figure=get_graph_1())),
                                html.Br(),
                                html.Div(dcc.Graph(figure=get_graph_2())),
                                html.Br(),
                                html.Div(dcc.Graph(figure=get_graph_3())),
                                ])


# Run the app
if __name__ == '__main__':
    app.run_server(host='0.0.0.0')
```
