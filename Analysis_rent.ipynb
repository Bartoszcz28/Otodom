```python
cdf = dat[["rent_price","m2","deposit","number_of_floors","room_number","floor"]]
cdf.head()
```

```python
avg_rent_price = cdf["rent_price"].astype("float").mean(axis=0)
print("Average rent_price: ", avg_rent_price)

avg_m2 = cdf["m2"].astype("float").mean(axis=0)
print("Average m2: ", avg_m2)

avg_deposit = cdf["deposit"].astype("float").mean(axis=0)
print("Average deposit: ", avg_deposit)

avg_number_of_floors = cdf["number_of_floors"].astype("float").mean(axis=0)
print("Average number_of_floors: ", avg_number_of_floors)

avg_room_number = cdf["room_number"].astype("float").mean(axis=0)
print("Average room_number: ", avg_room_number)

avg_floor = cdf["floor"].astype("float").mean(axis=0)
print("Average floor: ", avg_floor)

cdf["rent_price"].replace(np.nan, avg_rent_price, inplace=True)
cdf["m2"].replace(np.nan, avg_m2, inplace=True)
cdf["deposit"].replace(np.nan, avg_deposit, inplace=True)
cdf["number_of_floors"].replace(np.nan, avg_number_of_floors, inplace=True)
cdf["room_number"].replace(np.nan, avg_room_number, inplace=True)
cdf["floor"].replace(np.nan, avg_floor, inplace=True)
```

```python
avg_deposit = cdf["deposit"].astype("float").mean(axis=0)
print("Average deposit: ", avg_deposit)

cdf["deposit"].replace(np.nan, avg_deposit, inplace=True)
```

```python
avg_number_of_floors = cdf["number_of_floors"].astype("float").mean(axis=0)
print("Average number_of_floors: ", avg_number_of_floors)

cdf["number_of_floors"].replace(np.nan, avg_number_of_floors, inplace=True)
```

```python
avg_room_number = cdf["room_number"].astype("float").mean(axis=0)
print("Average room_number: ", avg_room_number)

cdf["room_number"].replace(np.nan, avg_room_number, inplace=True)
```

```python
avg_floor = cdf["floor"].astype("float").mean(axis=0)
print("Average floor: ", avg_floor)

cdf["floor"].replace(np.nan, avg_floor, inplace=True)
```

```python
cdf_t =  cdf[cdf.rent_price.between(0, 8000, inclusive=False)]
cdf_t.head()
```

```python
cdf_t.shape
```

```python
msk = np.random.rand(len(cdf_t)) < 0.8
train = cdf_t[msk]
test = cdf_t[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[["m2","deposit","number_of_floors","room_number","floor"]])
y = np.asanyarray(train[["rent_price"]])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[["m2","deposit","number_of_floors","room_number","floor"]])
x = np.asanyarray(test[["m2","deposit","number_of_floors","room_number","floor"]])
y = np.asanyarray(test[["rent_price"]])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
